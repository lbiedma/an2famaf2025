Demuestre que si $A$ es simétrica definida positiva, existen $L$ triangular inferior con unos en la diagonal y $D$ diagonal con entradas positivas tales que $A=LDL^T$.

Definimos $B$ como sigue:

$
B = G^TG = 
\begin{bmatrix}
g_{11} & 0^T\\
s & \hat{G}^T 
\end{bmatrix}
\begin{bmatrix}
g_{11} & s^T\\
0 & \hat{G} 
\end{bmatrix}
$

Por ser $B$ simétrica definida positiva. Supongamos que $G$ se puede escribir como sigue:

$G^T = L_B D_B^{1/2}$. Entonces definimos $D_B^{1/2}$ como sigue:


$D_B^{1/2} = 
\begin{bmatrix}
g_{11} & 0^T \\
0 & \hat{D}^{1/2}
\end{bmatrix}
=\begin{bmatrix}
g_{11} &     &    & \\
     & g_{22} &    & \\
     &      & \ddots & \\
     &      & & g_{nn}
\end{bmatrix}
$

diagonal con entradas positivas ($g_{ii} > 0$ para todo $i =1, 2, ..., n$). Y definimose $L_B$ como:

$
L_B = (D_B^{1/2})^{-1}G^T = 
\begin{bmatrix}
\dfrac{1}{g_{11}} & 0^T \\
0 & (\hat{D}^{1/2})^{-1}
\end{bmatrix}
\begin{bmatrix}
g_{11} & 0^T\\
s & \hat{G}^T 
\end{bmatrix}
=\begin{bmatrix}
1 & 0^T\\
g_{11}(\hat{D}^{1/2})^{-1} & (\hat{D}^{1/2})^{-1}\hat{G}^T 
\end{bmatrix}
$ 

Por hipotesis inductiva existe $\hat{L} = (\hat{D}^{1/2})^{-1}\hat{G}^T $

Luego, $L_B$ y $D_B^{1/2}$ existen y estan bien definidos.  

$B = G^T G = L_B D^{1/2}_B (D^{1/2}_B)^T L_B^T$

definimos $D_B = D^{1/2}_B (D^{1/2}_B)^T$

Luego, $B = L_B D_B L_B^T$
