Demuestre que si $A$ es simétrica definida positiva, existen $L$ triangular inferior con unos en la diagonal y $D$ diagonal con entradas positivas tales que $A=LDL^T$.

Sea $A \in R^{(n-1) \times (n-1)}$ es simetrica definida positiva y existe $L_A$ triangular inferior con 1 en la diagonal y $D_A$ diagonal con entradas positivas.   

$
B = \begin{bmatrix}
b_{11} & b^T\\
b & A 
\end{bmatrix} = 
G^TG = 
\begin{bmatrix}
g_{11} & 0^T\\
s & \hat{G}^T 
\end{bmatrix}
\begin{bmatrix}
g_{11} & s^T\\
0 & \hat{G} 
\end{bmatrix}
$

Por ser $B$ simétrica definida positiva

$G^T = L_B D_B^{1/2}$

$D_B^{1/2} = 
\begin{bmatrix}
g_{11} & 0^T \\
0 & \hat{D}^{1/2}
\end{bmatrix}
$

$
L_B = (D_B^{1/2})^{-1}G = 
\begin{bmatrix}
\dfrac{1}{g_{11}} & 0^T \\
0 & (\hat{D}^{1/2})^{-1}
\end{bmatrix}
\begin{bmatrix}
g_{11} & 0^T\\
s & \hat{G}^T 
\end{bmatrix}
=\begin{bmatrix}
1 & 0^T\\
g_{11}(\hat{D}^{1/2})^{-1} & (\hat{D}^{1/2})^{-1}\hat{G}^T 
\end{bmatrix}
$ 

Por hipotesis inductiva existe $\hat{L} = (\hat{D}^{1/2})^{-1}\hat{G}^T $

$B = G^T G = L_B D^{1/2}_B (D^{1/2}_B)^T L_B^T$

definimos $D_B = D^{1/2}_B (D^{1/2}_B)^T$
