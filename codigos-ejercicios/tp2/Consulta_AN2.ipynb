Sean $(x_i, y_i)$ numeros reales.

$$
x_i^2 + y_i^2+ax_i+by_i +c = 0  
$$

Despejando tenemos que,
$$
ax_i+by_i +c = -(x_i^2 + y_i^2)  
$$

Supongamos que tenemos $(x_i, y_i)$ $i=1,2,3. entonces tenemos el siguiente sistema lineal


$$
\begin{cases}
ax_1+by_1 +c &=& -(x_1^2 + y_1^2)\\
ax_2+by_2 +c &=& -(x_2^2 + y_2^2)\\
ax_3+by_3 +c &=& -(x_3^2 + y_3^2)
\end{cases}
$$

donde $a$, $b$ y $c$ son las incogintas a determinar.
