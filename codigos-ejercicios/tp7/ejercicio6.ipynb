Ejercicio 6.  Práctico 7

Para resolver $Ax=b$ con $A\in\R^{n\times n}$ simétrica definida positiva, considere el siguiente método iterativo: dado $x^k$ defina $x^{k+1}=x^k+d^k$ donde $d^k_i$ es solución de
$$
\mathop{\mathrm{minimizar}}_{d\in\R} \ f(x^k+de^i),
$$
con $f(x)=\frac{1}{2}x^TAx-b^Tx$ y $e^i$ el $i$-ésimo vector canónico. Demuestre que la sucesión generada $\{x^k\}$ coincide con la generada por el método de Jacobi.

dado $f$ y x^k+de^i tenemos que

$$
f(x^k+de^i) = \dfrac{1}{2}(x^k+de^i)^TA(x^k+de^i) - b^T(x^k+de^i)
$$

$$
f(x^k+de^i) = \dfrac{1}{2}((x^k)^TAx^T  +2(x^k)^TAde^i + d^2(e^i)^TAe^i) - b^Tde^i
$$

Entonces derivando f se obtiene

$$
f'(x^k+de^i) = \dfrac{1}{2}((2(x^k)^TAe^i + 2d(e^i)^TAe^i) - b^Te^i
$$

$$
f'(x^k+de^i) = (x^k)^TAe^i + d(e^i)^TAe^i - b^Te^i
$$

Como $f$ es una función cuadrática existe el mínimo y es único entonces el mínimo se alcanza cuando $f'(x) = 0$: 

$$
f'(x^k+de^i) = 0 = (x^k)^TAe^i + d(e^i)^TAe^i - b^Te^i
$$

$$ 
(x^k)^TAe^i + d(e^i)^TAe^i - b^Te^i = 0
$$

$$ 
\sum_{j=1}^{n} a_{ij}x_j + da_{ii} - b_i = 0
$$

$$ 
d = \dfrac{1}{a_{ii}}(b_i - \sum_{j=1}^{n} a_{ij}x_j)
$$

esto vale para todo $i =1, ... , n$ entonces $d$ se lo define como:

$$ 
d_i = \dfrac{1}{a_{ii}}(b_i - \sum_{j=1}^{n} a_{ij}x_j)
$$



Vamos a reemplazar en:
$$
x^{k+1} = x^k + d^k
$$
Veamos que pasa para cada elemento $x_{i}^{k+1}$
$$
x_i^{k+1} = x_i^k + d_i^k
$$

$$x_i^{k+1} = x_i^k +  \dfrac{1}{a_{ii}}(b_i - \sum_{j=1}^{n} a_{ij}x_j^k)$$

$$x_i^{k+1} = \dfrac{1}{a_{ii}}(b_i - \sum_{j=1, j\neq i}^{n} a_{ij}x_j^k)$$

esto vale para todo $i = 1, ..., n$
