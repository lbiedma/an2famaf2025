Ejercicio 4 del práctico 7

Sea $A\in\R^{n\times n}$ simétrica definida positiva $\{x^k\}$ generada por el método de Gastinel a partir de un $x^0$. Demuestre que
$$
 \|x^{k+1}-x^*\|_A\leq \left(1-\frac{1}{n\kappa_2(A)}\right)^{1/2} \|x^k-x^*\|_A,
$$
donde $Ax^*=b$. En particular $x^k\to x^*$.

Vamos a usar la siguiente igualdad 

$$
 \|x^{k+1}-x^*\|_A^2 = \left(1-\dfrac{(r^k)^T d^k}{(d^k)^TA d^k}\dfrac{(r^k)^T d^k}{(r^k)^TA^{-1} r^k}\right) \|x^k-x^*\|_A^2,
$$

la definicion de Gatinel es:

$$
d_i^k =
\begin{cases}
\frac{r_i^k}{|r_i^k|}&si\;r_i^k\neq 0\\
0 & r_i^k = 0
\end{cases}
$$

Para agilizar las cuentas llamamos $r = r^k$ y $d = d^k$

Empezamos por: 
$
\dfrac{r^T d}{d^TA d}\dfrac{r^T d}{r^TA^{-1} r}
$

Primera cuenta:

$$
r^Td = \sum_{i =1}^{n}r_i d_i = \sum_{i =1}^{n}|r_i| = || r ||_1 
$$

Segundo cuenta
$$
0 < d^TA d = |d^TA d| \leq ||d||_2||Ad||_2 \leq \sqrt{n} ||Ad||_2 \leq \sqrt{n} ||A||_2 ||d||_2 \leq n ||A||_2    
$$

$$\dfrac{1}{d^TAd} \geq \dfrac{1}{n||A||_2}$$

Tercera cuenta

$$
0<r^TA^{-1} r = |r^TA^{-1} r|= ||r||_2||A^{-1} r||_2 \leq ||r||_2||A^{-1}||_2 ||r||_2 = ||r||_2^2 ||A^{-1}||_2
$$

$$\dfrac{1}{r^TA^{-1}r} \geq \dfrac{1}{||r||^2_2||A^{-1}||_2}$$

Entonces:

$$
\dfrac{r^T d}{d^TA d}\dfrac{r^T d}{r^TA^{-1} r} \geq \dfrac{||r||_1}{n ||A||_2}\dfrac{||r||_1}{||r||_2^2 ||A^{-1}||_2} = \dfrac{||r||_1^2}{n \kappa_2(A)||r||_2^2} 
$$

Vamos a usar la siguiente desigualdad del ejercicio 1 del practico 3 de normas:

$$
||x||_2 \leq ||x||_1
$$

pero nos interesa:

$$
\dfrac{1}{||x||_2} \geq \dfrac{1}{||x||_1}
$$


entonces hacemos el siguiente reemplazo 


$$
\dfrac{||r||_1^2}{n \kappa_2(A)||r||_2^2} \geq \dfrac{||r||_1^2}{n \kappa_2(A)||r||_1^2} = \dfrac{1}{n \kappa_2(A)} 
$$

Por tanto,

$$
\dfrac{r^T d}{d^TA d}\dfrac{r^T d}{r^TA^{-1} r} \geq \dfrac{1}{n \kappa_2(A)}
$$

Para terminar

$$
 \|x^{k+1}-x^*\|_A^2 = \left(1-\dfrac{(r^k)^T d^k}{(d^k)^TA d^k}\dfrac{(r^k)^T d^k}{(r^k)^TA^{-1} r^k}\right) \|x^k-x^*\|_A^2 \leq \left(1-\dfrac{1}{n\kappa_2(A)}\right) \|x^k-x^*\|_A^2
$$
